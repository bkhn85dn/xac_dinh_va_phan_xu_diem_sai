```python
import numpy as np
from scipy.spatial import procrustes
import os

print("Bắt đầu thực hiện Procrustes Alignment...")

pose1_path = '/content/output/video1/X3D.npy'
pose2_path = '/content/output/video2/X3D.npy'

if not os.path.exists(pose1_path) or not os.path.exists(pose2_path):
    print("❌ Lỗi: Không tìm thấy file X3D.npy. Hãy đảm bảo Cell 6 đã chạy xong!")
else:
    pose1 = np.load(pose1_path)
    pose2 = np.load(pose2_path)

    # Đảm bảo 2 video khớp số lượng khung hình (lấy số nhỏ hơn)
    min_frames = min(pose1.shape[0], pose2.shape[0])

    aligned_pose1 = np.zeros_like(pose1[:min_frames])
    aligned_pose2 = np.zeros_like(pose2[:min_frames])
    disparities = []

    # Căn chỉnh từng frame một
    for i in range(min_frames):
        mtx1, mtx2, disparity = procrustes(pose1[i], pose2[i])
        aligned_pose1[i] = mtx1
        aligned_pose2[i] = mtx2
        disparities.append(disparity)

    # Lấy tọa độ trung bình cộng
    final_3d_pose = (aligned_pose1 + aligned_pose2) / 2.0

    # Lưu file nộp bài
    out_file = '/content/output/Final_MultiView_3D.npy'
    np.save(out_file, final_3d_pose)

    print(f"✅ Đã đồng bộ và căn chỉnh thành công {min_frames} frames.")
    print(f"✅ Độ lệch trung bình giữa 2 góc camera: {np.mean(disparities):.5f}")
    print(f"🎉 HOÀN TẤT! File kết quả cuối cùng nằm ở: {out_file}")
```
